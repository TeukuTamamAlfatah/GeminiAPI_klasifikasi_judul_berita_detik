```python
def clean_Content(Judul_Berita):
    Judul_Berita=re.sub('@[A-Za-z0-9]+','',Judul_Berita) #menghapus @
    Judul_Berita=re.sub(r'(?:\@|http?\://|https?\://|www)\S+','',Judul_Berita) #remove link http
    Judul_Berita=' '.join(Judul_Berita.split())
    return Judul_Berita
```

```python
def clean_label(Label):
    Label=re.sub('@[A-Za-z0-9]+','',Label) #menghapus @
    Label=re.sub(r'(?:\@|http?\://|https?\://|www)\S+','',Label) #remove link http
    Label=' '.join(Label.split())
    return Label
```

```python
def noise_removal(words):
    words=words.translate(str.maketrans('','',string.punctuation+string.digits)) #maketrans untuk memetakan atau mengganti nilai
    words=words.strip()
    return words
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data['Judul_clean'], data['Label'], test_size=0.2, random_state=42)
```

```python
sentences = [sentence.split() for sentence in X_train]
w2v_model = Word2Vec(sentences, vector_size=100, window=5, min_count=5, workers=4)
```

```python
def vectorize(sentence):
    words = sentence.split()
    words_vecs = [w2v_model.wv[word] for word in words if word in w2v_model.wv]
    if len(words_vecs) == 0:
        return np.zeros(100)
    words_vecs = np.array(words_vecs)
    return words_vecs.mean(axis=0)

X_train = np.array([vectorize(sentence) for sentence in X_train])
X_test = np.array([vectorize(sentence) for sentence in X_test])
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import MultinomialNB

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Transformasi data menggunakan MinMaxScaler
X_train_scaled = scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier


# Membuat instance dari model Logistic Regression
clf_logistic = LogisticRegression()
clf_logistic.fit(X_train, y_train)

# Membuat instance dari model Naive Bayes
clf_GaussianNB = OneVsRestClassifier(GaussianNB())
clf_GaussianNB.fit(X_train, y_train)

clf_naive_bayes = OneVsRestClassifier(MultinomialNB())
clf_naive_bayes.fit(X_train_scaled, y_train)

# Membuat instance dari model Support Vector Machine (SVM)
clf_svm = SVC()
clf_svm.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

def display_classification_reports(models, X_test, y_test):
    for name, model in models.items():
        print(f"Classification Report for {name}:")
        y_pred = model.predict(X_test)
        print(classification_report(y_test, y_pred))
        print("="*50)

# Membuat dictionary untuk menyimpan nama model dan modelnya
models = {
    "Logistic Regression": clf_logistic,
    "Naive Bayes": clf_naive_bayes,
    "SVM": clf_svm,
    "GaussianNB" : clf_GaussianNB
}

# Memanggil fungsi untuk menampilkan laporan klasifikasi untuk semua model
display_classification_reports(models, X_test, y_test)
```
